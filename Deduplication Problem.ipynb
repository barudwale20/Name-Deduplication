```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
deduplicated = []
```

```python

def uniquename(listx):
#    print(listx)
    
    ff = process.dedupe(listx, threshold=90)
    print(ff)
    for uname in ff:
        for x in range(len(dfsorted.index)):
            if(dfsorted['combname'].iloc[x]== uname):
                
                print(dfsorted['fn'].iloc[x]+' '+dfsorted['ln'].iloc[x] +' '+dfsorted['dob'].iloc[x]+' '+dfsorted['gn'].iloc[x])
                xx = dfsorted['fn'].iloc[x]
                y = dfsorted['ln'].iloc[x]
                z = dfsorted['gn'].iloc[x]
                a = dfsorted['dob'].iloc[x]
                deduplicated.append([xx,y,z,a])  
                break

```

```python
listx = []
count = 0
pos = 0
for x in range(len(dfsorted.index)):
    if(count<=len(dfsorted.index)-2):
        if((dfsorted['dob'].iloc[x]==dfsorted['dob'].iloc[x+1]) and (dfsorted['gn'].iloc[x]==dfsorted['gn'].iloc[x+1])):
            count=count+1
            name = dfsorted['fn'].iloc[x]+' '+dfsorted['ln'].iloc[x]
            listx.append(name)
        else:
            count=count+1
            listx.append(dfsorted['fn'].iloc[x]+' '+dfsorted['ln'].iloc[x])
            uniquename(listx)
            listx = []
    else:
        listx.append(dfsorted['fn'].iloc[x]+' '+dfsorted['ln'].iloc[x])
        uniquename(listx)
        listx = []
```
